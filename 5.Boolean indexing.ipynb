# Constructing munltiple boolen conditions
* Creating variable to hold each set of criteria independently
* Combining all the criteria together into single Boolean Series

* (and,or,not do not work in boolen indexing in pandas)
* (ther are replaced with "&","|","~" with wrapped in "()")

```python
criteria1 = df.imdb_score > 8
criteria2 = df.content_rating =='PG-13'
criteria3 = (df.title_year<2000) | (df.title_year>=2010)
```

```python
criteria2.head()
```

```python
criteria_final1 = criteria1 & criteria2 & criteria3
criteria_final1.head()
```

# Filtering with boolen indexing
* Creating two sets of movies with diffrent criteria
* Combining the two sets using the or operator
* Passing it to the indexing operator to filter data

```python
criteria_final1 = criteria1 & criteria2 & criteria3
```

```python
criteria_a1 = df.imdb_score < 5
criteria_a2 = df.content_rating =='R'
criteria_a3 = (df.title_year >= 2000) | (df.title_year <=2010)
criteria_final2 = criteria_a1 & criteria_a2 & criteria_a3
```

```python
#combining two criterias with or operator
criteria_final_all = criteria_final1 | criteria_final2
criteria_final_all.head()
```

```python
# simply pass it to the indexing opertor to filter the data
df[criteria_final_all].head()
```

```python
#to verify check the rows and columns with "loc" index
col = ['imdb_score','content_rating','title_year']
df_filter = df.loc[criteria_final_all,col]
df_filter.head(10)
```

## Replicating boolen indexing with index selection
* Using college dataset to select all institutions
* Comparing each of their performance against one another

```python
college = pd.read_csv(r'C:\Users\user\Desktop\Git_Python\New folder\data\college.csv')
# to specify which data will equal to that column
college[college['STABBR'] == 'TX'] .head()
```

```python
# changing the index name and label column
college2 = college.set_index('STABBR')
college2.loc['TX'].head()
```

```python
#lets compare the speed selection by %timeit
%timeit college[college['STABBR'] =='TX']
```

```python
%timeit college2.loc['TX']
```

```python
#lets do with indexing opertor
%timeit college2 = college.set_index('STABBR')
```

```python
#it is possible to select both multiple states with boolen and index
states = ['TX','CA','NY']
college[college['STABBR'].isin(states)]  #boolen selection
college2.loc[states].head()
```
