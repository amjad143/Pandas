# Producing Cartesian Products
* Adding two Series overlapping and non-identical indexes
* introduction:
* Two series or DataFrames operating with one another:
* indexes of each object align first before any operation begins
* This alignment always creates a Cartesian product

```python
#lets create two indexs that are diffrent and there values are same
#creating a Series 
s1 = pd.Series(index=list('aaab'),data=np.arange(4))
s1
```

```python
#creating a Second series
s2 = pd.Series(index=list('cababb'),data=np.arange(6))
s2
```

```python
#Addding both series is a Cartesian Product
s1 + s2
```

```python
#index contains same elements with same order example
s1 = pd.Series(index=list('aaabb'),data=np.arange(5))
s2 = pd.Series(index=list('aaabb'),data=np.arange(5))
s1+s2
```

```python
#lets change the order with indexs
s1 = pd.Series(index=list('aaabb'),data=np.arange(5))
s2 = pd.Series(index=list('bbaaa'),data=np.arange(5))
s1+s2
```

# Replicating idmax with Method Changing
* Slowly changing together basics method
* Finding all row index values that contain maximum column value
* Attempt an implementation of a built-in DataFrame method
* idmax is a challenging method to replicate

```python
college = pd.read_csv(r'C:\Users\user\Desktop\Git_Python\New folder\data\college.csv',index_col='INSTNM')
#now change this columns to be NUMERIC
cols = ['MD_EARN_WNE_P10','GRAD_DEBT_MDN_SUPP']
for col in cols:
       college[col] = pd.to_numeric(college[col],errors='coerce')
college_n = college.select_dtypes(include=[np.number])
criteria = college_n.nunique() == 2
binary_col = college_n.columns[criteria].tolist()
college_n = college_n.drop(labels=binary_col,axis='columns')
```

```python
#finding the maximum collumns of each method
college_n.max().head()
```

```python
#to test the column of each max
college_n.eq(college_n.max()).head()
```

```python
#to find out atleat one row which contains maximum value
has_row_max = college_n.eq(college_n.max()).any(axis='columns')
has_row_max.head()
```

```python
#finding the shape of the data
college_n.shape
#here it shows having 18 columns which means 18 true values
```

```python
#finding the row max
has_row_max.sum()
```

```python
#to check the 'idmax' returns the first current value remove ant method
#to check the rows that are maximum equal to it
college_n.eq(college_n.max()).cumsum()
#here first and the last three rows showing the output
```

```python
#if we want to change 'cumsum' method one more time
college_n.eq(college_n.max()).cumsum().cumsum()
```

```python
#now we can test the quality of each aginst ones
#to check the quality of the rows having atleast ONE
has_row_max2 = college_n.eq(college_n.max())\
                        .cumsum()\
                        .cumsum()\
                        .eq(1)\
                        .any(axis='columns')
has_row_max2.head()
#from this we can say that we dont have nomore true values
```

```python
#to find out the max2 row is true,use boolen index method
has_row_max2.sum()
```

```python
#to check wether they are same we found in the idmax method
idmax_col =has_row_max2[has_row_max2].index
idmax_col
```

```python
# to check wether they are same 
set(college_n.idxmax().unique()) == set(idmax_col)
```
