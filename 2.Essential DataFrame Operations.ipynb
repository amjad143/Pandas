# Ordering Column Names Sensibly
### Ordering columns by ProperGuidelines to columns 
#Guidelines to Order columns
#Classify each column as either discete or continuous
#Group commom columns with discrete and continuous columns
#Place the most important group of columns first

```python
df1=pd.read_csv(r'C:\Users\user\Desktop\Git_Python\New folder\data\movie.csv')
df1.head()
```

```python
df1.columns
```

```python
# Arranging all the columns in a order either discreate or continues
disc_core = ['movie_title','title_year','content_rating','genres']
disc_people = ['director_name','actor_1_name','actor_2_name','actor_3_name']
disc_others =['color','country','language','plot_keywords','movie_imdb_link']
cont_fb = ['director_facebook_likes','actor_1_facebook_likes','actor_2_facebook_likes','actor_3_facebook_likes',
           'cast_total_facebook_likes','movie_facebook_likes']
cont_finance = ['budget','gross']
cont_num_reviews = ['num_voted_users','num_user_for_reviews','num_critic_for_reviews']
cont_other = ['imdb_score','duration','aspect_ratio','facenumber_in_poster']
```

```python
new_col_order = disc_core+disc_people+disc_others+cont_fb+\
                cont_finance+cont_num_reviews+cont_other
```

```python
#here we get true we are checking entire data
set(df1.columns) == set(new_col_order)
```

```python
#re- arrange all the columns in order way
df1[new_col_order].head()
```
