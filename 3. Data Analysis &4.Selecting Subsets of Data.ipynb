# Selecting the Smallest of the Largest
* Finding 5 lowest movies from top 100 scoring movies
* Taking advantage of convenience method,nlrgest and nsmallest

```python
movie = pd.read_csv(r'C:\Users\user\Desktop\Git_Python\New folder\data\movie.csv')
movie2 = movie[['movie_title','imdb_score','budget']]
movie2.head()
```

```python
#selecting the largest movies rating
#syntax:(integer(row),column(index)).head()
movie2.nlargest(100,'imdb_score').head()
```

```python
#selecting the lowest budget movies or lowest index columns
movie2.nlargest(100,'imdb_score').nsmallest(5,'budget')
```

* Passing a list of column names to columns parameter is possible
* This breaks ties when there are duplicate values sharing

# Selecting the largest of each Group by sorting

```python
movie2=movie[['movie_title','title_year','imdb_score']]
movie2.sort_values('title_year',ascending=False).head()
```

```python
#to sort multiple columns at once we use a LIST
#use ascending parameter from the largest to smallest
movie3 = movie2.sort_values(['title_year','imdb_score'],ascending=False)
movie3.head()
```

##### Series and DataFrames: have complex data containers have multiple 'attributes'  using index operator  to selct '.iloc and .loc' are avalable these are called indexers
*.iloc: selects only by integer location;
* :works similarly to pyhton lists.
* .loc: selects only by index label
* : works python dictionaries        

```python
df = pd.read_csv(r'C:\Users\user\Desktop\Git_Python\New folder\data\college.csv',index_col='INSTNM')
city = df['CITY']
city.head()
# Selecting single column as series
```

```python
#The ILOC represent selects only integer values
city.iloc[3]
#it represents only sclar values
```

```python
#pass several integers to .ILOC
city.iloc[[10,20,30]]
```

```python
# to add multiple disjoint data in a particular manner
city.iloc[4:50:10] #[start:stop:step]
```

```python
# the loc will represents only sclar values which display indexs values
city.loc['Heritage Christian University']
#which selest only index labels
```

```python
#TO select several disjoint labels to add
np.random.seed(1)
labels = list(np.random.choice(city.index,4))
labels
#here we choose 4 index labels form the list randomly
```

```python
#to select several equally spaced portion data
city.loc['Alabama State University':'Reid State Technical College':10]
#here we can use the step integer for the selection
```

# speeding up Sclar Selection
* .iat: use integer location and must be passed two integers
* .at: index uses labels and must be passed an index and column label

```python
#To select a Sclar value for the particular index
df
cn = 'Texas A & M University-College Station'
df.loc[cn,'UGDS_WHITE']
```

```python
# we can acchive the same result with .at
df.at[cn,'UGDS_WHITE']
```

```python
#To find the speed of the both commands use magic function
%timeit df.loc[cn,'UGDS_WHITE']
```

```python
%timeit df.at[cn,'UGDS_WHITE']
```

```python
#finding the integer location of the both row and column number
row_num = df.index.get_loc(cn)
col_num = df.columns.get_loc('UGDS_WHITE')
row_num,col_num
```

```python
# to get the run time command for iloc
%timeit df.iloc[row_num,col_num]
```

```python
# to get the run time command for .iat
%timeit df.iat[row_num,col_num]
```
